```python
#import libraries for mapbox
import pandas as pd
import os
from pathlib import Path
from dotenv import load_dotenv
```

```python
#read in data
data_path = (Path("clean_data/employment_income_population_cleaned.csv"))
df = pd.read_csv(data_path)
df.head()
```

```python
#filter for 2019 only
df_2019 = df.loc[df['Year'] == 2019]
df_2019.head()
```

```python
#sort for largest 20 MSAs by Population
df_top_20 = df_2019.nlargest(20, 'Population')
df_top_20
```

```python
df_top_20.reset_index(inplace= True)
df_top_20
```

```python
df_top_20.drop(columns = ['index', 'Personal Income', 'Number Jobs'], inplace = True)
df_top_20
```

```python
df_top_20.to_csv('Top20_MSA.csv')
```
