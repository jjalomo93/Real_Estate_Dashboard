```python
#import libraries used for data cleaning
import pandas as pd
from pathlib import Path
```

```python
#pull raw csv data
csv = Path("raw_data/uscities_latitutude_longitude.csv")
df_lat_lon=pd.read_csv(csv)
df_lat_lon.head()
```

```python
#remove null values
df_lat_lon.dropna(inplace=True)
```

```python
#drop unneccessary columns
df_lat_lon.drop(df_lat_lon.columns.difference(['city','state_id', 'county_fips', 'lat', 'lng']), 1, inplace=True)
#df_lat_lon.drop(df_lat_lon.columns.difference(['county_fips', 'lat', 'lng']), 1, inplace=True)
```

```python
df_lat_lon.head()
```

```python
#combine city and state to create MSA identifier
df_lat_lon["cbsaname"] = df_lat_lon["city"] + ", " + df_lat_lon["state_id"]
df_lat_lon.head()
```

```python
#read in second csv file
csv_2 = Path("raw_data/cbsatocountycrosswalk.csv")
df_fips=pd.read_csv(csv_2)
df_fips.head()
```

```python
#remove null values
df_fips.dropna(inplace=True)
```

```python
#drop unneccessary columns
df_fips.drop(df_fips.columns.difference(['fipscounty','msaname', 'cbsa', 'cbsaname']), 1, inplace=True)
#df_fips.drop(df_fips.columns.difference(['fipscounty','cbsaname']), 1, inplace=True)
```

```python
df_fips.head()
```

```python
# #rename colunmn
# df_fips.rename(columns= {'fipscounty' : 'county_fips'}, inplace= True)
# df_fips.head()
```

```python
# df_lat_lon.set_index(keys= 'county_fips',inplace = True)
# df_fips.set_index(keys= 'county_fips',inplace = True)
```

```python
#combine dataframes
combined = df_lat_lon.merge(df_fips, how = "inner", on = "cbsaname")
combined
```

```python
#drop duplicates
combined.drop_duplicates(subset = "county_fips", inplace = True)
combined
```

```python
combined.reset_index(inplace= True)
combined
```

```python
combined.drop(columns=['index'], axis = 0,  inplace = True)
combined
```

```python
#create csv
combined.to_csv("clean_data/lat_lon_clean.csv", index=True, header=True)
```
