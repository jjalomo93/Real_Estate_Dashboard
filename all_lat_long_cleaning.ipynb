```python
#import libraries used for data cleaning
import pandas as pd
from pathlib import Path
```

```python
#pull latitude & longitude csv data
csv = Path("clean_data/lat_lon_clean.csv")
df_lat_lon=pd.read_csv(csv)
df_lat_lon.head()
```

```python
#drop uneccessary column
df_lat_lon.drop(['Unnamed: 0'], axis=1, inplace=True)
df_lat_lon.head()
```

```python
#drop uneccessary column
df_lat_lon.drop(['MSA'], axis=1, inplace=True)
df_lat_lon.head()
```

```python
#pull in all clean data
csv = Path("clean_data/all_top_20.csv")
df_all=pd.read_csv(csv)
df_all.head()
```

```python
#drop uneccessary column
df_all.drop(['Unnamed: 0'], axis=1, inplace=True)
df_all.head()
```

```python
#combine dataframes
combined = df_lat_lon.merge(df_all, how = "inner", on = "GeoFips")
combined
```

```python
#create csv
combined.to_csv("clean_data/latitude_longitude_top20.csv", index=True, header=True)
```

```python
#drop duplicates
combined.drop_duplicates(subset = "MSA", inplace = True)
combined.head()
```
