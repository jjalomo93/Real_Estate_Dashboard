```python
#import libraries used for data cleaning
import pandas as pd
from pathlib import Path
```

```python
#pull raw csv data
csv = Path("uscities_latitutude_longitude.csv")
df=pd.read_csv(csv)
df.head()
```

```python
#remove null values
df.dropna(inplace=True)
```

```python
#drop unneccessary columns
df.drop(df.columns.difference(['city','state_id', 'lat', 'lng']), 1, inplace=True)
```

```python
df.head()
```

```python
#combine city and state to create MSA identifier
df["MSA"] = df["city"] + ", " + df["state_id"]
```

```python
df.head()
```

```python
df.index= df['MSA']
```
