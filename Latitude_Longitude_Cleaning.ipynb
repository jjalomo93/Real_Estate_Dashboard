```python
#import libraries used for data cleaning
import pandas as pd
from pathlib import Path
```

```python
#pull latitude & longitude csv data
csv = Path('raw_data/cbsa_boundaries_2018.csv')
df=pd.read_csv(csv)
df.head()
```

```python
#drop unneccessary columns
df.drop(df.columns.difference(['Geo Point','NAMELSAD', 'GEOID']), 1, inplace=True)
df.head()
```

```python
#drop null values
df = df.dropna()
```

```python
#remove space from Geo Point column name
df.rename(columns= {'Geo Point' : 'GeoPoint'}, inplace= True)
df.head()
```

```python
#split Geo Point into separate latitude and longitude columns
#df.GeoPoint.str.split(expand=True)
df[['Latitude','Longitude']] = df.GeoPoint.str.split(expand=True)
df.head()
```

```python
#drop duplicates
df.drop_duplicates(subset = "NAMELSAD", inplace = True)
df.head()
```

```python
df.drop(columns=['GeoPoint'], axis = 0,  inplace = True)
df.head()
```

```python
#rename column
df.rename(columns= {'NAMELSAD' : 'MSA'}, inplace= True)
df.head()
```

```python
#rename column
df.rename(columns= {'GEOID' : 'GeoFips'}, inplace= True)
df.head()
```

```python
#remove commas from Latitude column
#df['Latitude'] = df['Latitude'].str.extract(',')
df.replace(',','', regex=True, inplace=True)
# #df['MSA'] = df['MSA'].str.replace('Micro Area', '')
df.head()
```
