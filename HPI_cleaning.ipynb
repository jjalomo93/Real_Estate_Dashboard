```python
# import libraries used for cleaning
import pandas as pd
from pathlib import Path
import datetime as dt
```

```python
# import the raw csv data from first csv and assign to a dataframe
csv = Path("FreddieMac_MSA_A-L.csv")
df1 = pd.read_csv(csv)
df1.head()
```

```python
# create date column in datetime format and remove NA values and unnecessary 2021 data
df1['Date'] = pd.date_range(start='1/1975', periods=len(df1), freq='M')
df1['Date'] = pd.to_datetime(df1['Date'])
df1.dropna(inplace=True)
df1.drop([552,553,554,555],inplace=True)
```

```python
# set index to date column and remove unecessary columns
df1.index= df1['Date']
df1.drop(columns=['Date','Month'],inplace=True)
```

```python
df1.tail()
```

```python
# change datatypes to float for resampling
df1 = df1.astype('float')
df1_yearly = df1.resample('Y').mean()
```

```python
# remove years prior to 2000 from dataset
df1_yearly = df1_yearly.iloc[25:]
df1_yearly.head()
```

________________________________________

```python
# import the raw csv data from second csv and assign to a dataframe
csv = Path("FreddieMac_MSA_M-Z.csv")
df2 = pd.read_csv(csv)
df2.head()
```

```python
# create date column in datetime format and remove NA values and unnecessary 2021 data
df2['Date'] = pd.date_range(start='1/1975', periods=len(df2), freq='M')
df2['Date'] = pd.to_datetime(df2['Date'])
df2 = df2.loc[:,~df2.columns.str.startswith('Unnam')]
df2.dropna(inplace=True)
df2.drop([552,553,554,555],inplace=True)
df2.tail()
```

```python
# set index to date column and remove unecessary columns
df2.index= df2['Date']
df2.drop(columns=['Date','Month'],inplace=True)
```

```python
df2.tail()
```

```python
# change datatypes to float for resampling
df2 = df2.astype('float')
df2_yearly = df2.resample('Y').mean()
```

```python
# remove years prior to 2000 from dataset
df2_yearly = df2_yearly.iloc[25:]
df2_yearly.head()
```

____________________________________________

```python
# merge the two cleaned dataframes together 
df_all = pd.concat([df1_yearly,df2_yearly], join='inner', axis='columns')
df_all.head()
```

```python
# create csv
df_all.to_csv("HPI_all.csv", index=True, header=True)
```

```python
# assign the dataset to a new dataframe and reset the index
df = df_all.reset_index()
```

```python
# transpose the data to show the locations as rows
df_transposed = df.transpose()
df_transposed.head()
```

```python
# reset the index
df_transposed.reset_index(inplace=True)
df_transposed.head()
```

```python
# set column headers as the first row
df_transposed.columns= df_transposed.iloc[0]
df_transposed.head()
```

```python
# drop first row
df_transposed.drop([0], inplace=True)
df_transposed.head()
```

```python
# melt the dataframe to show a single row for each location by year
df_clean = df_transposed.melt(
    id_vars=['Date'],
    var_name='Year',
    value_name='HPI'
)
df_clean.head()
```

```python
# rename the columns
df_clean = df_clean.rename(columns={'Date':'MSA'})
df_clean.head()
```

```python
# set index to GeoName
df_clean.set_index("MSA", inplace=True)
df_clean.head()
```

```python
# create csv
df_clean.to_csv("HPI_clean.csv", index=True, header=True)
```
