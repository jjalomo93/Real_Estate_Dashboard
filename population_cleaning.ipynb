```python
# import libraries used for data cleaning
import pandas as pd
from pathlib import Path
```

```python
# pull raw csv data
csv = Path("population_raw.csv")
```

```python
# check head and tail to see what rows need to be excluded
df = pd.read_csv(csv, header=4) # start at row 4 where columns start
df.tail(15)
```

```python
# remove last 10 columns that are not needed in data
df.drop([385, 386, 387, 388, 389, 390, 391, 392, 393, 394], inplace=True)
df.tail()
```

```python
# set reference column for GeoFips names
df_names = df[["GeoFips","GeoName"]]
df_names.set_index("GeoFips", inplace=True)
df_names.head()
```

```python
# drop GeoName column for easier cleaning of data
df.drop(columns=["GeoName"], inplace=True)
df.head()
```

```python
# Transpose data to show years as the row and GeoFips code as columns
df_transposed = df.transpose()
df_transposed.head()
```

```python
# reset the index
df_transposed.reset_index(inplace=True)
df_transposed.head(1)
```

```python
# set the column headers as the first row
df_transposed.columns = df_transposed.iloc[0]
df_transposed.head()
```

```python
# drop the duplicated columns
df_transposed.drop([0], inplace=True)
df_transposed.head()
```

```python
# melt the dataframe to show a single row for every GeoFips code by year
df_clean = df_transposed.melt(
    id_vars = ["GeoFips"],
    var_name = "Year",
    value_name = "Population"
)
df_clean.head()
```

```python
# rename the columns correctly
df_clean = df_clean.rename(columns={"GeoFips":"Year", "Year":"GeoFips"})
df_clean.head()
```

```python
# set the GeoFips column as the index
df_clean.set_index("GeoFips", inplace=True)
df_clean.head()
```

```python
# merge the cleaned df with the reference df to pull in GeoName
df_clean_names = pd.concat([df_clean, df_names], join="inner", axis="columns")
df_clean_names.head()
```

```python
# re-order the columns
df_clean_names = df_clean_names[["GeoName","Year","Population"]]
df_clean_names.head()
```

```python
# drop the United States combined data, not needed for project
df_clean_names.drop("00998", inplace=True)
df_clean_names.head()
```

```python
# create cleaned csv
df_clean_names.to_csv("population_cleaned.csv", index=True)
```
