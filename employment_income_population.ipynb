```python
import pandas as pd
from pathlib import Path
```

```python
emp_path = Path("employment_cleaned.csv")
emp = pd.read_csv(emp_path)
emp.drop(columns=["GeoName"],inplace=True)
emp.head()
```

```python
inc_path = Path("personal_income_cleaned.csv")
inc = pd.read_csv(inc_path)
inc.drop(columns=["GeoName"],inplace=True)
inc.head()
```

```python
pop_path = Path("population_cleaned.csv")
pop = pd.read_csv(pop_path)
pop.drop(columns=["GeoName"],inplace=True)
pop.head()
```

```python
msa_path = Path("msa_reference.csv")
msa = pd.read_csv(msa_path)
msa.head()
```

```python
combined = pop.merge(inc, on=["GeoFips","Year"], how="inner")
combined.head()
```

```python
combined2 = combined.merge(emp, on=["GeoFips","Year"], how="inner")
combined2.head()
```

```python
combined3 = combined2.merge(msa, on="GeoFips", how="inner")
combined3 = combined3[["GeoFips","MSA","Year","Population","Personal Income","Number Jobs"]]
combined3.head()
```

```python
combined3.to_csv("employment_income_population_cleaned.csv",index=False)
```
